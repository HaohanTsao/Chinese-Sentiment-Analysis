## 繁體

```python
# 設定資料夾路徑
folder_path = '/content/drive/MyDrive/python機器學習專案/raw_data'

# 指定要讀取的檔案擴展名
extension = '.csv'

# 要選取的欄位
selected_columns = ['content', 'sentiment']

# 使用 glob 匹配資料夾中符合條件的檔案
file_paths = glob.glob(f"{folder_path}/*{extension}")

# 儲存所有檔案的資料框清單
dataframes = []

# 逐一讀取符合條件的檔案
for file_path in file_paths:
    # 排除不想讀取的檔案條件
    if 'db2022_hw5_dcard_job' in file_path or 'db2022_hw5_dcard_food' in file_path:
        continue
    
    # 讀取 CSV 檔案並選取欄位
    try:
        df = pd.read_csv(file_path)
        df = df[selected_columns]
        dataframes.append(df)
    except pd.errors.ParserError as e:
        print(f"無法解析檔案：{file_path}")
        print(f"錯誤訊息：{str(e)}")
        continue

# 合併所有資料框
merged_df = pd.concat(dataframes, ignore_index=True)

# 顯示合併後的結果
print('合併後的資料框:')
print(merged_df)
```

```python
valid_sentiments = ['N', 'M', 'P']
merged_df = merged_df[merged_df['sentiment'].isin(valid_sentiments)]
```

```python
merged_df = merged_df.dropna().reset_index(drop=True)

merged_df
```

```python
merged_df['sentiment'] = merged_df['sentiment'].replace({'N': 0, 'M': 1, 'P': 2})
merged_df
```

```python
merged_df.to_csv('/content/drive/MyDrive/python機器學習專案/zh-TW_normalized_data.csv', index=False)
```

## 簡體


```python
# 設定資料夾路徑
folder_path = '/content/drive/MyDrive/python機器學習專案/zh-CN_data'

# 指定要讀取的檔案擴展名
extensions = ['.csv', '.tsv']

# 要排除的檔案名稱
exclude_files = ['test.csv', 'test.tsv']

# 儲存所有檔案的資料框清單
dataframes = []

# 逐一讀取符合條件的檔案
for extension in extensions:
    # 使用 glob 匹配資料夾中符合條件的檔案
    file_paths = glob.glob(f"{folder_path}/*{extension}")
    
    # 讀取 CSV 或 TSV 檔案並選取欄位
    for file_path in file_paths:
        # 排除不想讀取的檔案
        if any(exclude_file in file_path for exclude_file in exclude_files):
            continue
        
        # 判斷檔案類型
        if extension == '.csv':
            df = pd.read_csv(file_path)
            # 將欄位 'review' 改為 'content'
            df = df.rename(columns={'review': 'content'})
        elif extension == '.tsv':
            df = pd.read_csv(file_path, delimiter='\t')
            # 將欄位 'text_a' 改為 'content'
            df = df.rename(columns={'text_a': 'content'})
        
        # 將欄位 'label' 改為 'sentiment'
        df = df.rename(columns={'label': 'sentiment'})
        
        # 選取欄位 'content' 和 'sentiment'
        df = df[['content', 'sentiment']]
        
        dataframes.append(df)

# 合併所有資料框
cn_merged_df = pd.concat(dataframes, ignore_index=True)

# 顯示合併後的結果
print('合併後的資料框:')
print(cn_merged_df)
```

```python
cn_merged_df = cn_merged_df.dropna().reset_index(drop=True)

cn_merged_df
```

```python
cn_merged_df['sentiment'] = cn_merged_df['sentiment'].replace({0: 0,1: 2})
cn_merged_df
```

```python
cn_merged_df.to_csv('/content/drive/MyDrive/python機器學習專案/zh-CN_normalized_data.csv', index=False)
```
